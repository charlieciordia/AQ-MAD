```python
import pandas as pd
import folium
import h3
import numpy as np

# Crear un DataFrame aleatorio con puntos geográficos
np.random.seed(42)
n_points = 1000
latitudes = np.random.uniform(low=40, high=41, size=n_points)
longitudes = np.random.uniform(low=-74, high=-73, size=n_points)

data = {'LATITUD': latitudes, 'LONGITUD': longitudes}
df = pd.DataFrame(data)

# Asignar a cada punto una celda hexagonal de H3
resolution = 7  # Nivel de resolución H3 (ajusta según lo necesites)
df['H3'] = df.apply(lambda row: h3.geo_to_h3(row['LATITUD'], row['LONGITUD'], resolution), axis=1)

# Contar la cantidad de puntos por celda hexagonal
counts = df['H3'].value_counts().reset_index()
counts.columns = ['H3', 'count']

# Crear un mapa centrado en una ubicación
m = folium.Map(location=[40.75, -73.98], zoom_start=10)

# Agregar celdas hexagonales al mapa
for h3_index, count in counts.iterrows():
    lat, lng = h3.h3_to_geo_boundary(count['H3'])  # Obtener los límites de la celda hexagonal
    polygon = folium.Polygon(locations=zip(lat, lng), color='blue', fill_color='blue', fill_opacity=0.4)
    m.add_child(polygon)

# Mostrar el mapa
m
```

```python
import pandas as pd
import folium
import h3
import numpy as np

def get_h3_resolution(zoom):
    if zoom <= 5:
        return 4
    elif 5 < zoom <= 10:
        return 6
    elif 10 < zoom <= 15:
        return 8
    else:
        return 10

# Crear un DataFrame aleatorio con puntos geográficos
np.random.seed(42)
n_points = 1000
latitudes = np.random.uniform(low=40, high=41, size=n_points)
longitudes = np.random.uniform(low=-74, high=-73, size=n_points)

data = {'LATITUD': latitudes, 'LONGITUD': longitudes}
df = pd.DataFrame(data)

# Crear un mapa centrado en una ubicación
m = folium.Map(location=[40.75, -73.98], zoom_start=10)

# Agregar un controlador de eventos de zoom al mapa
def on_zoom(e):
    resolution = get_h3_resolution(e['target'].zoom)
    
    # Asignar a cada punto una celda hexagonal de H3
    df['H3'] = df.apply(lambda row: h3.geo_to_h3(row['LATITUD'], row['LONGITUD'], resolution), axis=1)

    # Contar la cantidad de puntos por celda hexagonal
    counts = df['H3'].value_counts().reset_index()
    counts.columns = ['H3', 'count']

    # Eliminar polígonos existentes del mapa
    for layer in m._children.values():
        if isinstance(layer, folium.features.Polygon):
            m.remove_layer(layer)

    # Agregar celdas hexagonales al mapa
    for h3_index, count in counts.iterrows():
        lat_lng = h3.h3_to_geo_boundary(count['H3'])  # Obtener los límites de la celda hexagonal
        polygon = folium.Polygon(locations=lat_lng, color='blue', fill_color='blue', fill_opacity=0.4)
        m.add_child(polygon)

m.on_zoom_end(on_zoom)

# Mostrar el mapa
m
```
