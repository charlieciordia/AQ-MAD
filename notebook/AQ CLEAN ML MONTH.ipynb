## MODELO ARIMA

```python
y_test = sol["NO2"][-30:]
y_train = sol["NO2"][:-30]
```

```python
from itertools import product
```

```python
ps = range(5)
ds = range(1,3)
qs = range(3)

parameters = list(product(ps,ds, qs))
print(parameters)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from tqdm import tqdm
```

```python
%%time
rmse = []
order = []
for p, d, q in tqdm(parameters):
    try:
        modelo=ARIMA(y_train, order=(p, d, q)).fit()
        pred = modelo.predict(start=len(y_train), end=len(sol) -1)
        error = np.sqrt(np.mean(mean_squared_error(y_test, pred)))
        rmse.append(error)
        order.append((p,d, q))
    except: 
        continue
```

```python
# veamos los resultados de todos los modelos
results = pd.DataFrame({"RMSE": rmse, "order": order})
results.sort_values(by="RMSE", inplace=True)
results.head()
```

```python
import itertools
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
```

```python
# Definir rangos de parmetros estacionales (P, D, Q) y trmino estacional (m)
ps = range(3)  # Ejemplo: 0 a 2
qs = range(3)  # Ejemplo: 0 a 2
Ps = range(3)  # Ejemplo: 0 a 2
Qs = range(3)  # Ejemplo: 0 a 2
m = 12  # El trmino estacional, por ejemplo, para datos mensuales

parametros = list(itertools.product(ps, qs, Ps, Qs))

rmse = []
orden = []

for p, q, P, Q in parametros:
    try:
        modelo = SARIMAX(y_train, order=(p, 0, q), seasonal_order=(P, 0, Q, m)).fit()
        pred = modelo.predict(start=len(y_train), end=len(plaza_esp) - 1)
        error = np.sqrt(mean_squared_error(y_test, pred))
        rmse.append(error)
        orden.append((p, q, P, Q))
    except:
        continue  # Ignora combinaciones que arrojen errores

# Encuentra la combinacin con el RMSE ms bajo
mejor_orden = orden[np.argmin(rmse)]
mejor_rmse = np.min(rmse)

print("Mejor orden:", mejor_orden)
print("Menor RMSE:", mejor_rmse)
```
