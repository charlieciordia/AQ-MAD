```python
lst_estaciones = [8, 11, 17, 18, 24, 27, 35, 36, 38, 39, 40, 47, 48, 49, 50, 54, 55, 56, 57, 58, 59, 60]
```

```python
dataframes_por_estacion = {}

for estacion in lst_estaciones:
    nombre_df = f'df_{estacion}'  # Nombre del DataFrame
    
    df_estacion = df[df["ESTACION"] == estacion].copy()  # Filtrar el DataFrame por cada estación
    
    # Guardar el DataFrame con el nombre específico en el entorno local
    locals()[nombre_df] = df_estacion
    
    # Guardar el DataFrame en el diccionario para referencia adicional si es necesario
    dataframes_por_estacion[nombre_df] = df_estacion
    
    # Imprimir información sobre el DataFrame creado
    print(f"DataFrame {nombre_df} creado. Filas: {len(df_estacion)}, Columnas: {len(df_estacion.columns)}")
```
