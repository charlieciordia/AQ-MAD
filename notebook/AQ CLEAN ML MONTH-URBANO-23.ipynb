## MODELO ARIMA

```python
y_test = sol["NO2"][-30:]
y_train = sol["NO2"][:-30]
```

```python
from itertools import product
```

```python
ps = range(5)
ds = range(1,3)
qs = range(3)

parameters = list(product(ps,ds, qs))
print(parameters)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from tqdm import tqdm
```

```python
%%time
rmse = []
order = []
for p, d, q in tqdm(parameters):
    try:
        modelo=ARIMA(y_train, order=(p, d, q)).fit()
        pred = modelo.predict(start=len(y_train), end=len(sol) -1)
        error = np.sqrt(np.mean(mean_squared_error(y_test, pred)))
        rmse.append(error)
        order.append((p,d, q))
    except: 
        continue
```

```python
# veamos los resultados de todos los modelos
results = pd.DataFrame({"RMSE": rmse, "order": order})
results.sort_values(by="RMSE", inplace=True)
results.head()
```

```python
modelo = ARIMA(y_train, order = (4,1,2) ).fit()
pred = modelo.predict(start = len(y_train), end = len(sol) -1)
```

## SARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
%%time

modelo=SARIMAX(endog=y_train, 
               exog=None,
               #order=(4, 1, 2),  # modelo arima
               seasonal_order=(4, 1, 2, 12) # modelo sarima, teniendo en cuenta la estacionalidad
              ).fit(disp=False)
                     
pred=modelo.predict(len(y_train), len(sol)-1)  

error=(pred-y_test).abs().sum()/len(pred) 
print ('Error mae: ', error)


res=pd.DataFrame({'real':y_test, 'pred':pred, 'error':pred-y_test})
res
```

```python
plt.plot(res["pred"], c = "r")
plt.plot(res["real"], c = "black");
```
