```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Escalar los datos de NO2 para mejorar la visualizaci√≥n
scaler = MinMaxScaler()
scaled_NO2 = scaler.fit_transform(df["NO2"].values.reshape(-1, 1))

# Calcular la media m√≥vil con una ventana de 12 meses
rolling_mean = pd.Series(scaled_NO2.squeeze()).rolling(window=12).mean()

# Crear el gr√°fico
plt.figure(figsize=(10, 6))

# Graficar la serie temporal de NO2 y la media m√≥vil
plt.plot(scaled_NO2, label='NO2')
plt.plot(rolling_mean, color='red', label='Media m√≥vil (12 meses)')

# Agregar etiquetas y t√≠tulo
plt.xlabel('Tiempo')
plt.ylabel('Valor escalado de NO2')
plt.title('Serie Temporal de NO2 con Media M√≥vil')

# Mostrar leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

```python
y_test = df["NO2"][-30:]
y_train = df["NO2"][:-30]
```

```python
from itertools import product
```

```python
ps = range(15)
qs = range(3)

parametros = list(product(ps, qs))
print(parametros)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = []
orden = []

for p, q in parametros:
    modelo = ARIMA(y_train, order = (p, 0, q)).fit()
    pred = modelo.predict(start = len(y_train), end = len(df) -1)
    error = np.sqrt(np.mean(mean_squared_error(y_test, pred)))
    rmse.append(error)
    orden.append((p, q))
```

```python
results = pd.DataFrame({"RMSE": rmse, "order": orden})
results.sort_values(by ="RMSE", ascending = True).head()
```
