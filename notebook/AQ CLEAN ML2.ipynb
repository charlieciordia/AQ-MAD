```python
y_test = plaza_esp["NO2"][-30:]
y_train = plaza_esp["NO2"][:-30]
```

```python
plaza_esp.shape
```

```python
from itertools import product
```

```python
ps = range(6)
qs = range(4)

parametros = list(product(ps, qs))
print(parametros)
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = []
orden = []

for p, q in parametros:
    modelo = ARIMA(y_train, order = (p, 0, q)).fit()
    pred = modelo.predict(start = len(y_train), end = len(plaza_esp) -1)
    error = np.sqrt(np.mean(mean_squared_error(y_test, pred)))
    rmse.append(error)
    orden.append((p, q))
```

```python
results = pd.DataFrame({"RMSE": rmse, "order": orden})
results.sort_values(by ="RMSE", ascending = True).head()
```
