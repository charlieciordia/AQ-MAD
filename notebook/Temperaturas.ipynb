### Temperatura Madrid

```python
import pandas as pd
```

```python
temp = pd.read_csv("../data/aemet/2012-2023.csv")
```

```python
temp.head()
```

```python
temp.drop(["index", "Provincia", "Media_prec", "Media_sol", "Media_tmin", "Media_tmax"], axis=1, inplace=True)
```

```python
temp.head()
```

```python
temp.rename(columns={'Media_tmed': 'T_media'}, inplace=True)
```

```python
temp.isnull().sum()
```

```python
temp['periodo'] = pd.to_datetime(temp['Periodo'])

# Filtra los registros que no pertenecen al a√±o 2023
temp = temp[temp['periodo'].dt.year != 2023]
temp = temp[temp['periodo'].dt.year != 2012]
```

```python
temp.shape
```

```python
temp.dtypes
```

```python
temp.index = pd.to_datetime(temp["Periodo"])
temp.head()
```

```python
temp.index.freq = pd.infer_freq(temp.index)
```

```python
temp.drop(["Periodo", "periodo"],  inplace = True, axis = 1)
```

```python
temp.head()
```

```python
temp.shape
```

```python
temp.to_csv('temp_mad_clean.csv', index=True)
```
