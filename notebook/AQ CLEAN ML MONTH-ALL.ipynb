```python
df.drop(['MUNICIPIO', 'PUNTO_MUESTREO'], axis=1, inplace = True)
```

```python
df_no2 = df[df["MAGNITUD"] == 8]
df_no2.head()
```

```python
df_no2.drop(['MAGNITUD'], axis=1, inplace = True)
```

```python
df_clean = pd.melt(df_no2, id_vars=['ESTACION', 'ANO', 'MES'], var_name='DIA', value_name='NO2')
df_clean.head()
```

```python
df_clean['DIA'] = df_clean['DIA'].str[1:].astype(int)
```

```python
monthly_avg = df_clean.groupby(['ESTACION','ANO','MES'])['NO2'].agg(lambda x: x.mean())
monthly_avg
```

```python
df_monthly_avg = monthly_avg.reset_index()
df_monthly_avg.head()
```

### Moda de todas las estaciones de MADRID

```python
# Agrupar por MES, ANO y NO2 y contar las ocurrencias de cada valor de NO2
monthly_mode = df_monthly_avg.groupby(['MES', 'ANO', 'NO2']).size().reset_index(name='count')

# Encontrar el valor de NO2 con el recuento mximo para cada MES y AO
monthly_mode = monthly_mode.loc[monthly_mode.groupby(['MES', 'ANO'])['count'].idxmax()]

monthly_mode
```

```python
monthly_mode.drop("count", axis=1, inplace=True)
```

```python
monthly_mode['month'] = monthly_mode['ANO'].astype(str) + '-' + monthly_mode['MES'].astype(str).str.zfill(2)
```

```python
monthly_mode.head()
```

```python
monthly_mode.drop(['ANO', 'MES'], axis=1, inplace=True)
```

```python
monthly_mode.head()
```

```python
monthly_mode.shape
```

```python
# Asegurar que la columna de ndice sea de tipo fecha
monthly_mode.index = pd.to_datetime(monthly_mode.index)

# Obtener el rango de fechas entre la fecha mnima y mxima en tus datos
rango_fechas = pd.date_range(start=monthly_mode.index.min(), end=monthly_mode.index.max(), freq='MS')  # 'MS' para frecuencia mensual

# Verificar si todas las fechas estn presentes en el ndice
fechas_faltantes = rango_fechas[~rango_fechas.isin(monthly_mode.index)]

if fechas_faltantes.empty:
    print("No faltan fechas en tus registros.")
else:
    print("Faltan fechas en tus registros:", fechas_faltantes)
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Escalar los datos de NO2 para mejorar la visualizacin
scaler = MinMaxScaler()
scaled_NO2 = scaler.fit_transform(monthly_mode["NO2"].values.reshape(-1, 1))

# Calcular la media mvil con una ventana de 12 meses
rolling_mean = pd.Series(scaled_NO2.squeeze()).rolling(window=12).mean()

# Crear el grfico
plt.figure(figsize=(10, 6))

# Graficar la serie temporal de NO2 y la media mvil
plt.plot(scaled_NO2, label='NO2')
plt.plot(rolling_mean, color='red', label='Media mvil (12 meses)')

# Agregar etiquetas y ttulo
plt.xlabel('Tiempo')
plt.ylabel('Valor escalado de NO2')
plt.title('Serie Temporal de NO2 con Media Mvil')

# Mostrar leyenda
plt.legend()

# Mostrar el grfico
plt.show()
```
